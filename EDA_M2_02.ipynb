```python
import pandas as pd
import numpy as np
import missingno as msno
```

# Step 1: Load the dataset

```python
df = pd.read_csv('Crime_02.csv')
df.head()
```

```python
df.shape
```

```python
#Check for null values
df.isnull()
```

```python
df.isnull().sum()
```

```python
n=df.select_dtypes(include=np.number).columns
len(n)
```

```python
#finding number of unique values
for i in n:
  print("No. of unique values in ",i,"are: ",df.loc[:,i].nunique())
```

# Step 2: Check if there are any NA values

```python
any_na = df.isna().any().any()
print(f"Any NA values? {any_na}")
```

# Step 3: Finding missing summaries

```python
na_summary = df.isna().sum()
print("Missing values summary:\n", na_summary)
```

# Step 4: Total number of NA values in the dataset

```python
total_na = df.isna().sum().sum()
print(f"Total number of NA values: {total_na}")
```

# Step 5: Proportion of missing and complete values

```python
prop_missing = df.isna().mean()
prop_complete = 1 - prop_missing
print(f"Proportion of missing values:\n{prop_missing}")
print(f"Proportion of complete values:\n{prop_complete}")
```

# Step 6: Row-wise deletion of missing data

```python
df_clean = df.dropna()
print("Data after row-wise deletion of missing values:\n", df_clean)
```

#Step 7: Removing Duplicates

```python
df_deduplicated = df.drop_duplicates()
print("Data after deduplication:\n", df_deduplicated)
```
